```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the California housing dataset
housing_data = fetch_california_housing()
df_housing = pd.DataFrame(housing_data.data, columns=housing_data.feature_names)
y_housing = housing_data.target

# 1. Scatter plot of Median Income vs. Housing Price
plt.figure(figsize=(12, 6))
plt.scatter(df_housing['MedInc'], y_housing, alpha=0.5)
plt.title('Median Income vs. Housing Price')
plt.xlabel('Median Income')
plt.ylabel('Housing Price')
plt.grid(True)
plt.show()

# 2. Histogram of Housing Prices
plt.figure(figsize=(12, 6))
plt.hist(y_housing, bins=50)
plt.title('Distribution of Housing Prices')
plt.xlabel('Housing Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 3. Box plot of Median House Age
plt.figure(figsize=(12, 6))
sns.boxplot(df_housing['HouseAge'])
plt.title('Box plot of Median House Age')
plt.grid(True)
plt.show()

# 4. Scatter plot of Latitude and Longitude to visualize the geographical distribution of data
plt.figure(figsize=(12, 6))
plt.scatter(df_housing['Longitude'], df_housing['Latitude'], alpha=0.5, c=y_housing, cmap='viridis')
plt.colorbar(label='Housing Price')
plt.title('Geographical Distribution of Data')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()

# 5. Correlation heatmap of features
plt.figure(figsize=(12, 6))
correlation_matrix_housing = df_housing.corr()
sns.heatmap(correlation_matrix_housing, annot=True)
plt.title('Correlation Heatmap for California Housing Data')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load the California housing dataset
housing_data = fetch_california_housing()
df_housing = pd.DataFrame(housing_data.data, columns=housing_data.feature_names)
y_housing = housing_data.target

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(df_housing, y_housing, test_size=0.2, random_state=42)

# Training a RandomForestRegressor model
model_california = RandomForestRegressor(n_estimators=100, random_state=42)
model_california.fit(X_train, y_train)
y_pred_california = model_california.predict(X_test)

# Model Evaluation: Mean Squared Error
mse_california = mean_squared_error(y_test, y_pred_california)

# Visualizations
# ... [Same visualizations as provided above]

print(f"Mean Squared Error for California Housing Predictions: {mse_california:.2f}")
```
